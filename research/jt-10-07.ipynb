```python
import matplotlib.pyplot as plt
```

```python
df['Income Period'].head(40)
```

```python
def parse_income_period_with_leading_zeros(value):
    str_val = str(value).zfill(8)  # Add leading zeros to make it 8 digits
    year = "20" + str_val[:2]
    month = str_val[2:4]
    day = str_val[4:6]
    return f"{year}-{month}-{day}"

# Extract only the rows where the 'Income Period' is 8 digits long
df_8_digit = df[df['Income Period'].apply(lambda x: len(str(x)) == 8)]
pct = len(df_8_digit) / len(df)
print(f"pct worked {pct}")

# Parsing these rows as dates
df_8_digit['Parsed Date'] = df_8_digit['Income Period'].apply(parse_income_period_with_leading_zeros)

# Convert to datetime format for histogram plotting
df_8_digit['Parsed Date'] = pd.to_datetime(df_8_digit['Parsed Date'], errors='coerce')

# Plotting a histogram of the parsed dates

plt.figure(figsize=(10, 6))
df_8_digit['Parsed Date'].hist(bins=30)
plt.title('Histogram of Parsed Income Period Dates (8-digit values)')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
